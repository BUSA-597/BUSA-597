```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#reading the data
dataframe = pd.read_csv("apple_annual_revenue.csv")

#Checking the first rows of the data 
dataframe.head()
```

```python
#checking for null values in columns
dataframe.isnull().sum()
```

```python
dataframe.describe()
```

```python
dataframe.hist()
```

```python
#checking the distribution of the close price
plt.figure(figsize = (20, 15))
sns.displot(dataframe['Revenue(US $)'])
```

```python
dataframe.info()
```

```python
#revenue over time from 2010 t0 2021
dataframe.groupby(['Year'])[['Revenue(US $)']].mean().plot(kind="line",
                                    figsize=(15,10), title='Apple Revenue by year from 2012',
                                    ylabel='The Average revenue')
```

```python
#reading the second data
df = pd.read_csv("apple_two_months_stock_price.csv")

#Checking the first rows of the data
df.head()
```

```python
df.info()
```

```python
#Checking for the sum of null values in each column
df.isnull().sum()
```

```python
# getting Year, Month and WeekDay column from Date column
df['Year'] = pd.to_datetime(df['Date'], errors='coerce').dt.year
df['Month'] = pd.to_datetime(df['Date'], errors='coerce').dt.strftime('%b')
df['Week Day'] = pd.to_datetime(df['Date'], errors='coerce').dt.day_name()
df
```

```python
#Two month stock price chart
df.groupby(['Month'])[['Close']].mean().plot(kind="line",
                                    figsize=(15,10), title='Stock Price by Month',
                                    ylabel='The Average Stock Price')
```
