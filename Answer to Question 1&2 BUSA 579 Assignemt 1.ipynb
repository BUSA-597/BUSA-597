```python
# Removing Outliers

# Max = 75% + 1.5* (75% - 25%)

maximum = 2.247650e+07 + 1.5* (2.247650e+07 - 4.197385e+05)

new_data = data[data[ "2022 Population"] <= maximum] 

new_data.plot.box()
```

```python
# Comparing the frequency of pupulation in 2022, 2020, and 2015 across continent on the chart?
data.groupby(['Continent'])[['2022 Population', '2020 Population', '2015 Population']].sum().sort_values(
                                    ['2022 Population', '2020 Population', '2015 Population'], 
                                    ascending=False).plot(kind="bar",
                                    figsize=(15,10), title='Total population across Continent',
                                    ylabel='Sum of population')


#table display
data.groupby(['Continent'])[['2022 Population', '2020 Population', '2015 Population']].sum().sort_values(
                                    ['2022 Population', '2020 Population', '2015 Population'], 
                                    ascending=False)
```

Conclusion: The total population is much higher in Asia than other continent and there is a rise in population from 2015 to 2022, while Ocenia has the lowest total population with no significant changes from 2015 to 2020
